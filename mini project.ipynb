```python
plt.figure(figsize=(6, 4))
total_values.plot(kind='bar', color=['royalblue', 'orange'])
plt.title('Distribution of Target Values')
plt.xlabel('Target')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
sns.countplot(x='sex',hue='target',data=data)
plt.xticks([1,0],['Male','Female'])
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```

```python
sns.distplot(data['age'], bins=20)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Count occurrences of each chest pain type
counts = data['cp'].value_counts()

# Define labels for the x-axis ticks
labels = ["typical angina", "atypical angina", "non-anginal pain", "asymptomatic"]

# Define colors for each chest pain type
colors = ['skyblue', 'salmon', 'lightgreen', 'orange']

# Plotting
plt.bar(counts.index, counts.values, color=colors)
plt.title('Distribution of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.xticks([0, 1, 2, 3], labels, rotation=75)
plt.show()
```

```python
sns.countplot(x='cp',hue='target',data=data)
plt.legend(labels=['No-Disease','Diseases'])
plt.show()
```

```python
sns.countplot(x='fbs',hue='target',data=data)
plt.legend(labels=['No-Disease','Diseases'])
plt.show()
```
